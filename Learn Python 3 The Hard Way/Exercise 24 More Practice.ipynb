## Common Student Questions

- Q: What do you call the variable jelly_beans but the name breans later? A: That's part of how a function works. Remember that inside the function the variable is temporary. When you return it then it can be assigned to a variable for later. I'm just making a new variable named breans to hold the turn value.
- Q: What do you mean by reading the code backward? A: Start at the last line. Compare that line in your file to do the same line in mine. Once it's exactly the same, move up to the next line. Do this until you get to the first line of the file.
