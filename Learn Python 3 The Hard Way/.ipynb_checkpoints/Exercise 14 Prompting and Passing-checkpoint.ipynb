# Prompting and Passing

Let's do one exercise that uses argv and input together to ask the user something specific. You will need this for the next exercise where you learn to read and write files. In this exercise we'll use input slightly different by having a simple > prompt. This is similar to games like Zork or Adventure.


We make a variable, prompt, that is set to the prompt we want, and we give that to input instead of typing it over and over. Now if we want to make the prompt something else, we just change it in this one spot and rerun the script. Very handy.

##### SELF NOTE: The script needed to be ran in the command prompt. Running it in Powershell will not work! 

## What You Should See

When you run this, remember that you have to give the script your name for the argv arguments.
