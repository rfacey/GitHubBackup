## Common Student Questions
- Q: How can there be 10 different ways to run a function? A: Believe it or not, there's a theoretically infinite number of ways to call any function. See how creative you get with functions, variables, and input from a user.

- Q: Is there a way to analyze what this function is doing so I can understand it better? There are many different ways, but try putting an English comment above each line describing what the line does. Another trick is to read the code out loud. Yet another is to print the code out and draw on the paper with pictures and comments showing what's going on. 

- Q: What if I want to ask the user for the numbers of cheese and crackers? A: You need to use int() to convert what you get from input().

- Q: Does making the variable amount_of_cheese change the variable cheese_count in the function? A: No, those variables are separate and live outside the function. They are then passed to the function, and temporary versions are made just for the function's run. When the function exits, these temporary variables go away and everything keeps working. Keep going in the book, and this should become clearer.

- Q: Is it bad to have global variables (like amount_of_cheese) with the same name as function variables? A: Yes, since then you're not quite sure which one you're talking about. But sometimes nexessity means you have to use the same name, or you might do it on accident. Just avoid it whenever you can.

- Q: Is there a limit to the number of arguments a function can have? It depends on the version of Python and the computer you're on, but it is fairly large. The practical limit though is about five arguemnts before the function becomes annoying to use.

- Q: Can you call a function within a function? Yes, you'll make a game that does this later in the book.
